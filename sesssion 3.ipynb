```python
""" 2.Write a Python program that takes a user's input for the price of a Zomato order as a string, converts it to
 a float using type casting, adds 18% GST, and prints the final bill amount. """

price = input("Enter price of a Zomato order: ")
price = float(price)
gst = price * 0.18
final_bill = price + gst

print(final_bill)
```

```python
""" 3.Given a list of strings representing product prices from Flipkart, like ['199.99', '299.50', '150'],
convert all to floats and calculate the total cart value. """

prices = ['199.99', '299.50', '150']

price1 = float(prices[0])
price2 = float(prices[1])
price3 = float(prices[2])

total = price1 + price2 + price3

print("Tota Value =", total)
```

```python
"""4. Build a function is_discount_applicable(order_amount) that takes a float and returns True if the amount is greater than 500, otherwise False.
 Print the result for order amounts 450 and 750."""

def is_discount_applicable(order_amount):
    if order_amount > 500:
        return True
    else:
        return False
    
print(is_discount_applicable(450))
print(is_discount_applicable(750))
```

```python
"""5.You received a dataset of ratings as strings from Spotify: ['4.5', '3.0', '5', '4.2']. Use type casting to convert these to floats, then find and print the highest rating.
<br><br><em><strong>Hint:</strong> Use the float() function inside a loop or list comprehension.</em>"""

ratings = ['4.5', '3.0', '5', '4.2']

float_ratings = [float(rating) for rating in ratings]

print("Highest Rating:", max(float_ratings))
```
